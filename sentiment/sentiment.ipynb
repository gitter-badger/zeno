```python
from transformers import pipeline
import pandas as pd
from datasets import load_dataset
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
dataset = load_dataset("amazon_polarity")
```

```python
classifier("We are very happy to show you the ðŸ¤— Transformers library.")
```

```python
dfs = dataset['test'].to_pandas().sample(1000).reset_index(drop=True)
```

```python
dfs.head()
```

```python
out = classifier(dfs['content'].to_list())
```

```python
out
```

```python
dfs['out'] = out
```

```python
dfs['out'] = dfs['out'].apply(lambda x: (1 if x['label'] == 'POSITIVE' else 0))
```

```python
dfs.head()
```

```python
dfs.to_csv('amazon_pred.csv')
```
